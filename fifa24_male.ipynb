## Limpieza datos

```python
df = df.drop(columns=['Unnamed: 0', 'URL', 'Gender'])
df.columns
```

## Análisis por equipos

```python
# DataFrame con los jugadores pertenecientes a los 30 equipos con mejor media (Se hace la consulta a 31 porque OL Reign solo contiene 1 jugador)
df_teams = df[df['Club'].isin(df.groupby('Club')['Overall'].mean().sort_values(ascending=False).head(31).index)]

# Eliminar los jugadores pertenecientes al equipo OL Reign
df_teams = df_teams[df_teams['Club'] != 'OL Reign']

df_teams['Club'].value_counts()
```

```python
data = df_teams.groupby('Club')['Overall'].mean().sort_values(ascending=False)
data = data.reset_index()

plt.figure(figsize=(15, 8))
sns.barplot(data=data, x='Overall', y='Club', palette='viridis', hue='Overall')
plt.title('Top 30 equipos con mejor media de rating')
plt.xlabel('Media de Rating')
plt.ylabel('Club')
plt.show()
```

```python
data = df_teams.groupby('Club')['Age'].mean().sort_values(ascending=False)
data = data.reset_index()

plt.figure(figsize=(15, 8))
sns.barplot(data=data, x='Age', y='Club', palette='viridis', hue='Age')
plt.title('Media de edad mayor a menor de los 30 equipos')
plt.xlabel('Media de Edad')
plt.ylabel('Club')
plt.show()
```

```python
# DataFrame con los jugadores pertenecientes a los 30 equipos con más nacionalidades y sus respectivas nacionalidades
data = df_teams.groupby('Club')['Nation'].value_counts().sort_values(ascending=False).head(30).reset_index(name='counts')
data
```

```python
# Gráfica de los 30 equipos con más nacionalidades y sus respectivas nacionalidades
plt.figure(figsize=(15, 8))
sns.barplot(data=data, x='counts', y='Club', palette='viridis', hue='counts')
plt.title('Top 30 equipos con más nacionalidades')
plt.xlabel('Número de Nacionalidades')
plt.ylabel('Club')
plt.show()
```

```python
data = df.groupby('Club')['Nation'].nunique().sort_values(ascending=False).head(20)
data = data.reset_index()

plt.figure(figsize=(15, 8))
sns.barplot(data=data, x='Nation', y='Club', palette='viridis', hue='Nation')
plt.title('Top 20 equipos con más variedad de nacionalidades')
plt.xlabel('Número de Nacionalidades')
plt.ylabel('Club')
plt.show()
```

## Análisis de Jugadores

### Delanteros

```python
df_delanteros = df[df['Position'].isin(['ST', 'CF', 'LW', 'RW'])]
df_delanteros['Position'].value_counts()
```

```python
top30_delanteros = df_delanteros.sort_values(by='Overall', ascending=False).head(30)

plt.figure(figsize=(15, 8))
sns.barplot(data=top30_delanteros, x='Overall', y='Name', palette='viridis', hue='Overall')
plt.title('Top 30 delanteros con mejor media de rating')
plt.xlabel('Media de Rating')
plt.ylabel('')
plt.show()
```

```python
# Función para convertir nombres completos a iniciales
def convertir_a_iniciales(nombre_completo):
    partes = nombre_completo.split()
    if len(partes) > 1:
        return f"{partes[0][0]}. {partes[-1]}"
    return partes[0] 

top30_delanteros['Iniciales'] = top30_delanteros['Name'].apply(convertir_a_iniciales)
```

```python
fig, ax = plt.subplots(2, 3, figsize=(20, 10)) 

sns.barplot(data=top30_delanteros, x='Pace', y='Iniciales', palette='viridis', ax=ax[0, 0], hue='Pace')
ax[0, 0].set_title('Velocidad (Pace)')
ax[0, 0].set_ylabel('')
sns.barplot(data=top30_delanteros, x='Shooting', y='Iniciales', palette='viridis', ax=ax[0, 1], hue='Shooting')
ax[0, 1].set_title('Disparo (Shooting)')
ax[0, 1].set_ylabel('')
sns.barplot(data=top30_delanteros, x='Passing', y='Iniciales', palette='viridis', ax=ax[0, 2], hue='Passing')
ax[0, 2].set_title('Pases (Passing)')
ax[0, 2].set_ylabel('')
sns.barplot(data=top30_delanteros, x='Dribbling', y='Iniciales', palette='viridis', ax=ax[1, 0], hue='Dribbling')
ax[1, 0].set_title('Regate (Dribbling)')
ax[1, 0].set_ylabel('')
sns.barplot(data=top30_delanteros, x='Defending', y='Iniciales', palette='viridis', ax=ax[1, 1], hue='Defending')
ax[1, 1].set_title('Defensa (Defending)')
ax[1, 1].set_ylabel('')
sns.barplot(data=top30_delanteros, x='Physicality', y='Iniciales', palette='viridis', ax=ax[1, 2], hue='Physicality')
ax[1, 2].set_title('Físico (Physicality)')
ax[1, 2].set_ylabel('')

fig.suptitle('Top 30 delanteros con mejor media de estadísticas por atributo', fontsize=16)
plt.tight_layout()
plt.subplots_adjust(top=0.92)
plt.show()
```

```python
top30_delanteros['New Overall'] = (top30_delanteros['Pace'] + top30_delanteros['Shooting'] + top30_delanteros['Passing'] + top30_delanteros['Dribbling'] + top30_delanteros['Defending'] + top30_delanteros['Physicality']) / 6

df_melted = top30_delanteros.melt(id_vars=['Iniciales'], value_vars=['Overall', 'New Overall'], var_name='Tipo', value_name='Valor')
plt.figure(figsize=(15, 8))
sns.barplot(data=df_melted, x='Valor', y='Iniciales', palette='viridis', hue='Tipo')
plt.title('Comparativa de Overall y New Overall')
plt.xlabel('Valor')
plt.ylabel('')
plt.legend(title='Tipo', bbox_to_anchor=(1, 1), loc='upper left')
plt.tight_layout()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=top30_delanteros, x='Weak foot', y='Skill moves', hue='Overall', size='Overall', sizes=(20, 500), alpha=0.7)
plt.title('Weak Foot vs Skill Moves')
plt.xlabel('Weak Foot')
plt.ylabel('Skill Moves')
plt.legend(title='Overall', bbox_to_anchor=(1, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
