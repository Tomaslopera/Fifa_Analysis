## Limpieza datos

```python
df = df.drop(columns=['Unnamed: 0', 'URL', 'Gender'])
df.columns
```

## Análisis por equipos

```python
# DataFrame con los jugadores pertenecientes a los 30 equipos con mejor media (Se hace la consulta a 31 porque OL Reign solo contiene 1 jugador)
df_teams = df[df['Club'].isin(df.groupby('Club')['Overall'].mean().sort_values(ascending=False).head(31).index)]

# Eliminar los jugadores pertenecientes al equipo OL Reign
df_teams = df_teams[df_teams['Club'] != 'OL Reign']

df_teams['Club'].value_counts()
```

```python
data = df_teams.groupby('Club')['Overall'].mean().sort_values(ascending=False)
data = data.reset_index()

plt.figure(figsize=(15, 8))
sns.barplot(data=data, x='Overall', y='Club', palette='viridis', hue='Overall')
plt.title('Top 30 equipos con mejor media de rating')
plt.xlabel('Media de Rating')
plt.ylabel('Club')
plt.show()
```

```python
data = df_teams.groupby('Club')['Age'].mean().sort_values(ascending=False)
data = data.reset_index()

plt.figure(figsize=(15, 8))
sns.barplot(data=data, x='Age', y='Club', palette='viridis', hue='Age')
plt.title('Media de edad mayor a menor de los 30 equipos')
plt.xlabel('Media de Edad')
plt.ylabel('Club')
plt.show()
```

```python
# DataFrame con los jugadores pertenecientes a los 30 equipos con más nacionalidades y sus respectivas nacionalidades
data = df_teams.groupby('Club')['Nation'].value_counts().sort_values(ascending=False).head(30).reset_index(name='counts')
data
```

```python
# Gráfica de los 30 equipos con más nacionalidades y sus respectivas nacionalidades
plt.figure(figsize=(15, 8))
sns.barplot(data=data, x='counts', y='Club', palette='viridis', hue='counts')
plt.title('Top 30 equipos con más nacionalidades')
plt.xlabel('Número de Nacionalidades')
plt.ylabel('Club')
plt.show()
```

```python
data = df.groupby('Club')['Nation'].nunique().sort_values(ascending=False).head(20)
data = data.reset_index()

plt.figure(figsize=(15, 8))
sns.barplot(data=data, x='Nation', y='Club', palette='viridis', hue='Nation')
plt.title('Top 20 equipos con más variedad de nacionalidades')
plt.xlabel('Número de Nacionalidades')
plt.ylabel('Club')
plt.show()
```
