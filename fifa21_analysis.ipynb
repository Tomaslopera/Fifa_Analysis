```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Preprocesamiento de datos

```python
df = pd.read_csv('../Fifa Data Analysis/Datasets/2021/fifa21 raw data v2.csv')
df
```

```python
df.columns
```

```python
df.info()
```

#### Eliminación de columnas innecesarias

```python
df = df.drop(columns=['ID', 'photoUrl', 'playerUrl'])
df
```

#### Modificación de nombre OVERALL

```python
df['Overall'] = df['↓OVA']
df = df.drop(columns=['↓OVA'])
df
```

#### Eliminación de columna LongName

```python
df = df.drop(columns=['LongName'])
df
```

#### Limpieza de datos Club

```python
df['Club'] = df['Club'].str.replace('\n\n\n\n', '')
df
```

#### Modificación de datos Contract

```python
df['Contract'] = df['Contract'].str.replace('~', '-')
df
```

#### Modificación de formatos Height y Weight

```python
df['Height'] = df['Height'].str.replace('cm', '')
df['Weight'] = df['Weight'].str.replace('kg', '')
df
```

```python
df = df.rename(columns={'Height': 'Height (cm)', 'Weight': 'Weight (kg)'})
df
```

```python
df['Height (cm)'] = pd.to_numeric(df['Height (cm)'], errors='coerce')
df['Weight (kg)'] = pd.to_numeric(df['Weight (kg)'], errors='coerce')
df
```

#### Identificación

```python
df.columns
```

```python
df['POT']
```

#### Conversión de Joined Date a DATETIME

```python
df = df.rename(columns={'Joined': 'Joined Date'})
df['Joined Date'] = pd.to_datetime(df['Joined Date'], errors='coerce')
df['Joined Date']
```

#### Conversión de Loan Date a DATETIME

```python
df['Loan Date End'] = pd.to_datetime(df['Loan Date End'], errors='coerce')
df['Loan Date End']
```

#### Conversión de Value a INT

```python
df['Value'] = df['Value'].str.replace('€', '')
if df['Value'].str.contains('.').any():
    df['Value'] = df['Value'].str.replace('M', '00000')
    df['Value'] = df['Value'].str.replace('K', '000')
else:
    df['Value'] = df['Value'].str.replace('M', '000000')
    df['Value'] = df['Value'].str.replace('K', '000')

df['Value'] = df['Value'].str.replace('.', '')

df = df.rename(columns={'Value': 'Value (€)'})
df['Value (€)'] = pd.to_numeric(df['Value (€)'], errors='coerce')
df['Value (€)']
```

#### Conversión de Wage a INT

```python
df['Wage'] = df['Wage'].str.replace('€', '')
df['Wage'] = df['Wage'].str.replace('K', '000')
df['Wage'] = pd.to_numeric(df['Wage'], errors='coerce')
df = df.rename(columns={'Wage': 'Wage (€)'})
df['Wage (€)']
```

#### Conversión de Release Clause a INT

```python
df['Release Clause'] = df['Release Clause'].str.replace('€', '')
if df['Release Clause'].str.contains('.').any():
    df['Release Clause'] = df['Release Clause'].str.replace('M', '00000')
    df['Release Clause'] = df['Release Clause'].str.replace('K', '000')
else:
    df['Release Clause'] = df['Release Clause'].str.replace('M', '000000')
    df['Release Clause'] = df['Release Clause'].str.replace('K', '000')

df['Release Clause'] = df['Release Clause'].str.replace('.', '')

df = df.rename(columns={'Release Clause': 'Release Clause (€)'})
df['Release Clause (€)'] = pd.to_numeric(df['Release Clause (€)'], errors='coerce')
df['Release Clause (€)']
```

#### Modificación estrellas

```python
df['W/F'] = df['W/F'].str.replace('★', '')
df['SM'] = df['SM'].str.replace(' ★', '')
df['IR'] = df['IR'].str.replace(' ★', '')

df = df.rename(columns={'W/F': 'W/F (★)', 'SM': 'SM (★)', 'IR': 'IR (★)'})
df
```

#### Modificación A/W y D/W

```python
df['A/W'] = df['A/W'].replace({'Low': 'L', 'Medium': 'M', 'High': 'H'})
df['D/W'] = df['D/W'].replace({'Low': 'L', 'Medium': 'M', 'High': 'H'})
df
```

#### Eliminar atributos (Ya se relizó un Análisis de atributos en fifa 24)

```python
df_cleaned = df.copy()
df_cleaned.columns
```

```python
df_cleaned = df_cleaned.drop(columns=['Attacking', 'Crossing', 'Finishing',
       'Heading Accuracy', 'Short Passing', 'Volleys', 'Skill', 'Dribbling',
       'Curve', 'FK Accuracy', 'Long Passing', 'Ball Control', 'Movement',
       'Acceleration', 'Sprint Speed', 'Agility', 'Reactions', 'Balance',
       'Power', 'Shot Power', 'Jumping', 'Stamina', 'Strength', 'Long Shots',
       'Mentality', 'Aggression', 'Interceptions', 'Positioning', 'Vision',
       'Penalties', 'Composure', 'Defending', 'Marking', 'Standing Tackle',
       'Sliding Tackle', 'Goalkeeping', 'GK Diving', 'GK Handling',
       'GK Kicking', 'GK Positioning', 'GK Reflexes', 'Total Stats',
       'Base Stats', 'A/W', 'D/W', 'Hits', 'POT', 'BOV', 'IR (★)', 'PAC', 'SHO',
       'PAS', 'DRI', 'DEF', 'PHY'])
```

## Análisis

```python
df_cleaned.info()
```

```python
data = df_cleaned.groupby('Club')['Value (€)'].sum().sort_values(ascending=False).reset_index().head(20)

top20 = df_cleaned[df_cleaned['Club'].isin(data['Club'])]

plt.figure(figsize=(12, 8))
sns.barplot(data=data, x='Value (€)', y='Club', palette='viridis', hue='Value (€)')
plt.title('Top 20 Equipos por el valor total de sus jugadores')
plt.xlabel('Total Player Value (€)')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
data = top20.groupby('Club')['Height (cm)'].mean().sort_values(ascending=False).reset_index().head(20)

plt.figure(figsize=(12, 8))
sns.barplot(data=data, x='Height (cm)', y='Club', palette='viridis', hue='Height (cm)')
plt.title('Altura media de los jugadores de los 20 equipos con mayor valor total')
plt.xlabel('Average Height (cm)')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
data = top20.groupby('Club')['Weight (kg)'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(12, 8))
sns.barplot(data=data, x='Weight (kg)', y='Club', palette='viridis', hue='Weight (kg)')
plt.title('Peso medio de los jugadores de los 20 equipos con mayor valor total')
plt.xlabel('Average Weight (kg)')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
data = df_cleaned.groupby('Club')['Release Clause (€)'].sum().sort_values(ascending=False).reset_index().head(20)

plt.figure(figsize=(12, 8))
sns.barplot(data=data, x='Release Clause (€)', y='Club', palette='viridis', hue='Release Clause (€)')
plt.title('Top 20 Equipos por la cláusula de rescisión total de sus jugadores')
plt.xlabel('Total Release Clause (€)')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
relacion = top20.melt(id_vars=['Club'], value_vars=['Value (€)', 'Release Clause (€)'], var_name='Tipo', value_name='Valor')

plt.figure(figsize=(12, 8))
sns.barplot(data=relacion, x='Valor', y='Club', hue='Tipo', palette='viridis')
plt.title('Relación entre el valor total de los jugadores y la cláusula de rescisión')
plt.xlabel('Valor (€)')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

#### Artibutos

```python
df_delanteros = df[df['Positions'].str.contains('ST|CF|RW|LW|RF|LF')]
top30_delanteros = df_delanteros.sort_values(by='Overall', ascending=False).head(30)

plt.figure(figsize=(12, 8))
sns.barplot(data=top30_delanteros, x='Overall', y='Name', palette='viridis', hue='Overall')
plt.title('Top 30 Delanteros por su valoración')
plt.xlabel('Overall')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2, 3, figsize=(20, 10)) 

sns.barplot(data=top30_delanteros, x='PAC', y='Name', palette='viridis', ax=ax[0, 0], hue='PAC')
ax[0, 0].set_title('Velocidad (Pace)')
ax[0, 0].set_ylabel('')
sns.barplot(data=top30_delanteros, x='SHO', y='Name', palette='viridis', ax=ax[0, 1], hue='SHO')
ax[0, 1].set_title('Disparo (Shooting)')
ax[0, 1].set_ylabel('')
sns.barplot(data=top30_delanteros, x='PAS', y='Name', palette='viridis', ax=ax[0, 2], hue='PAS')
ax[0, 2].set_title('Pases (Passing)')
ax[0, 2].set_ylabel('')
sns.barplot(data=top30_delanteros, x='DRI', y='Name', palette='viridis', ax=ax[1, 0], hue='DRI')
ax[1, 0].set_title('Regate (Dribbling)')
ax[1, 0].set_ylabel('')
sns.barplot(data=top30_delanteros, x='DEF', y='Name', palette='viridis', ax=ax[1, 1], hue='DEF')
ax[1, 1].set_title('Defensa (Defending)')
ax[1, 1].set_ylabel('')
sns.barplot(data=top30_delanteros, x='PHY', y='Name', palette='viridis', ax=ax[1, 2], hue='PHY')
ax[1, 2].set_title('Físico (Physicality)')
ax[1, 2].set_ylabel('')

fig.suptitle('Top 30 delanteros con mejor media de estadísticas por atributo', fontsize=16)
plt.tight_layout()
plt.subplots_adjust(top=0.92)
plt.show()
```
